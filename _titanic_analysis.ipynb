```python
import pandas  as pd
url='https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df=pd.read_csv(url)
print(df)
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python

df = df.drop(columns=['Cabin'], errors='ignore')
print(df.head())
```

```python
df.dropna(subset=['Embarked']  ,inplace=True)
```

```python
df.isnull().sum()
```
